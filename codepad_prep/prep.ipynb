## Dataset Parsing

```python
import os
import pandas as pd

dataset_dir = "./assignment_datasets"

def parse(directory : str) -> pd.DataFrame:
    dfs = []
    # for root, dirs, files in os.walk(directory):
    for file in os.listdir(directory):
        #for file in files:
        if (file.split('.')[-1] == "csv"):
            dfs.append(pd.read_csv(os.path.join(directory, file), index_col="geography"))
    return pd.concat(dfs)

df = parse(dataset_dir)
df.head(5)
```
