# Decision Trees



Decision trees recursively partition the dataset into two. At each node, they decide a feature and value to split on. Each root node permits a distribution over output classes:

$$ \pi_{q}(R_{\alpha}) = \frac{\sum_{x_{(i)}} I(x_{(i)} \in R_{\alpha} \cap y^{(i)} = c_{q})}{\sum_{x_{(i)}} I(x_{(i)} \in R_{\alpha})} $$

The proportion of elements in $R_{\alpha}$ in class $c_{q}$. The output probabilities of our tree is:

$$ \pi(R_{\alpha}) = \left(\pi_{1}(R_{\alpha}), \pi_{2}(R_{\alpha}), \dots, \pi_{q}(R_{\alpha}) \right) $$

We aim to reduce the within partition variation of $\pi$. We do this by considering one of two metrics:

Gini-index: $\sum_{c = 1}^{q} \pi_{c}(R_{\alpha})(1 - \pi_{c}(R_{\alpha}))$

Cross-Entropy $-\sum_{c=1}^{q} \pi_{c}(R_{\alpha})\log(\pi_{c}(R_{\alpha}))$



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 101, endpoint=True)
plt.plot(x, x * (1-x), label="Gini Index")
plt.plot(x, -x * np.log(x + 10e-5), label="Cross-Entropy")
plt.legend()
plt.show()
```
